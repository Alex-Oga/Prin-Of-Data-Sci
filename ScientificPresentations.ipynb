```python
dataBeer = (beer,)
dataWater = (water,)
rng = np.random.default_rng()
```

```python
std_sample = np.std(dataBeer)  # the sample statistic
print(std_sample)
std_sample = np.std(dataWater)  # the sample statistic
print(std_sample)
```

```python
dist = norm(loc=2, scale=4)  # our "unknown" distribution
std_true = dist.std()      # the true value of the statistic
print(std_true)
```

```python
bootstrapBeer_ci = bootstrap(dataBeer, np.std, confidence_level=0.9,
                random_state=rng)
bootstrapWater_ci = bootstrap(dataWater, np.std, confidence_level=0.9,
                random_state=rng)

print(bootstrapBeer_ci.confidence_interval)
print(bootstrapWater_ci.confidence_interval)
```

```python
print(bootstrapBeer_ci)
print(bootstrapWater_ci)
```

```python
plt.hist(bootstrapWater_ci.bootstrap_distribution)
plt.hist(bootstrapBeer_ci.bootstrap_distribution)
```

```python
a = 0
for i in bootstrapWater_ci.bootstrap_distribution:
    if i > np.std(dataWater):
        a += 1
print(a)
b = 0
for k in bootstrapBeer_ci.bootstrap_distribution:
    if k > np.std(dataBeer):
        b += 1
print(b)
print("p-value of Water:", a/len(bootstrapWater_ci.bootstrap_distribution))
print("p-value of Beer:", b/len(bootstrapBeer_ci.bootstrap_distribution))
```

```python
def estimate_p_value_permutation_test(a, b, n_sims=100_000):
  d = np.mean(a) - np.mean(b)
  k = len(a)
  whole = np.append(a, b)
  count = 0
  diffs = []
  for i in range(n_sims):
      diff = np.mean(whole[:k]) - np.mean(whole[k:])
      diffs.append(diff)
      if abs(diff) >= abs(d):
          count += 1
      np.random.shuffle(whole)
  p_value = count/n_sims
  return(p_value, diffs)
```

```python
p_value, diffs = estimate_p_value_permutation_test(pipe1,pipe2)
p_value
```

```python
plt.hist(diffs, density=True) 
plt.vlines(np.mean(pipe1) - np.mean(pipe2), 0, 0.25, colors = ['red']) 
plt.show()
```

```python
def estimate_p_value_bootstrap2(a, b, n_sims=10_000):
  """
  Given two groups of data a and b, estimate p_value that their means can differ by at least as measured, 
  assuming the null hypothesis that a and b come from the same distribution

  returns p_value and list of possible difference in means.
  """
  d = np.mean(a) - np.mean(b)
  la, lb = len(a), len(b)
  whole = np.append(a, b)

  count = 0
  diffs = []

  for i in range(n_sims):
    boot_a = np.random.choice(whole,la)
    boot_b = np.random.choice(whole,lb)
    diff = np.mean(boot_a) - np.mean(boot_b)
    diffs.append(diff)
    if abs(diff) >= abs(d):
      count += 1

  p_value = count/n_sims
  return(p_value, diffs)
```

```python
p_value, diffs = estimate_p_value_bootstrap2(pipe1,pipe2)
p_value
```

```python
plt.hist(diffs, density=True) 
plt.vlines(np.mean(pipe1) - np.mean(pipe2), 0, 0.25, colors = ['red']) 
plt.show()
```
