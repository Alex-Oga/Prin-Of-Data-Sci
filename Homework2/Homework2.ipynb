```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.DataFrame( {-2:[8.0, 10.0], -1:[9.0, 10.3], 0:[9.1, 12.2], 1:[10.2, 12.6], 2:[10.4, 13.9]}, ['A', 'B'])
df
```

```python
X1 = df.iloc[0, :]
```

```python
X2 = df.iloc[1, :]
```

```python
Y = [-2, -1, 0, 1, 2]
```

```python
plt.scatter(X1, Y)
plt.scatter(X2, Y)
```

```python
sns.regplot(x=X1,y=Y,ci=None,color ='red');
sns.regplot(x=X2,y=Y,ci=None,color ='blue');
```

```python
X3 = X1.values.reshape(-1,1)
```

```python
ols = linear_model.LinearRegression()
model = ols.fit(X3, Y)
response = model.predict(X3)
```

```python
r2 = model.score(X3, Y)
```

```python
plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(X3, response, color='k', label='Regression model')
ax.scatter(X3, Y, edgecolor='k', facecolor='grey', alpha=0.7, label='Sample data')

fig.tight_layout()
```

```python
X4 = X2.values.reshape(-1,1)
```

```python
ols = linear_model.LinearRegression()
model = ols.fit(X4, Y)
response = model.predict(X4)
```

```python
r2 = model.score(X4, Y)
```
