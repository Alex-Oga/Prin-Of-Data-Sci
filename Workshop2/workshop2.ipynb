```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Gender.csv', sep=';')
```

```python
df.plot()
```

```python
df.plot(kind='scatter',x='height',y='male')
```

```python
X = df['height'].values
Y = df['male'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
```

```python
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)
```

```python
print("Coefficients")
print(m, c)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100
 
x = np.linspace(min_x, max_x, 1000)
y = c + m * x
 
plt.plot(x, y, color='#58b970', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
 
plt.xlabel('Height')
plt.ylabel('Male')
plt.axvline(np.median(X), 0, 1, label='Cut off')
plt.legend()
plt.show()
```

```python
print("Cut-Off Height")
print(np.median(X))
```
